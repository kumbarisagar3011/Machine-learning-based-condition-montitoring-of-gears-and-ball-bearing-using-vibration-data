WHen training with the SVM Radial Basis Function, two parameters are considered **'C'** and **'Gamma'**
A low **C** makes the decision surface smooth, while high **C** aims at classifying all the training examples correctly.
**gamma** defines how much influence s single training example has. The larger **gamma** is, closer other examples must be affected.
Use GridSeardCV fo fing optimal **C** and **gamma**
